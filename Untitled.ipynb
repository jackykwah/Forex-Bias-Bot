```python
import pandas as pd
import numpy as np
import os

# --- 1. DATA LOADER ---
def load_xau_file(file_path):
    df = pd.read_csv(file_path, sep=';')
    df.columns = [col.strip().lower() for col in df.columns]
    df['date'] = pd.to_datetime(df['date'], format='%Y.%m.%d %H:%M')
    df.set_index('date', inplace=True)
    df.sort_index(inplace=True)
    return df

# --- 2. DEFENSIVE POI LOGIC (BCC Rule Included) ---
def identify_poi(df, timeframe_name):
    df = df.copy()
    
    # Force Booleans and fill NaNs to prevent the "Float" TypeError
    df['is_bullish'] = (df['close'] > df['open']).fillna(False).astype(bool)
    df['body_size'] = (df['close'] - df['open']).abs()
    df['avg_body'] = df['body_size'].rolling(window=5).mean().fillna(0)
    df['is_impulsive'] = (df['body_size'] > (df['avg_body'] * 2)).astype(bool)

    # Shifted boolean to avoid the ~ operator error
    prev_bullish = df['is_bullish'].shift(1).fillna(False).astype(bool)
    curr_bullish = df['is_bullish'].astype(bool)
    is_impulsive = df['is_impulsive'].astype(bool)

    # New Zone Detection
    df['new_supply'] = np.where((prev_bullish == True) & (curr_bullish == False) & (is_impulsive == True), 
                                df['high'].shift(1), np.nan)
    df['new_demand'] = np.where((prev_bullish == False) & (curr_bullish == True) & (is_impulsive == True), 
                                df['low'].shift(1), np.nan)

    # --- BCC Invalidation Loop (Timeframe Specific) ---
    active_s = np.nan
    active_d = np.nan
    s_history = []
    d_history = []

    for i in range(len(df)):
        row = df.iloc[i]
        # Update zones if new ones appear
        if pd.notna(row['new_supply']): active_s = row['new_supply']
        if pd.notna(row['new_demand']): active_d = row['new_demand']
        
        # BCC RULE: Only invalidate if the candle BODY closes past the zone
        if pd.notna(active_s) and row['close'] > active_s: active_s = np.nan
        if pd.notna(active_d) and row['close'] < active_d: active_d = np.nan
        
        s_history.append(active_s)
        d_history.append(active_d)

    df[f'active_supply_{timeframe_name}'] = s_history
    df[f'active_demand_{timeframe_name}'] = d_history
    return df

# --- 3. THE "STATE MACHINE" EXECUTION (ChoCh + BOS) ---
def run_backtest_pro(df, initial_balance=10000, risk_per_trade=0.01):
    balance = initial_balance
    pos = None # 'long', 'short', or None
    history = []
    
    # State tracking for confirmation sequence
    tapped = False
    choch_found = False
    
    for i in range(1, len(df)):
        row = df.iloc[i]
        prev = df.iloc[i-1]
        
        # --- A. EXIT LOGIC ---
        if pos:
            if (pos == 'short' and row['high'] >= sl) or (pos == 'long' and row['low'] <= sl):
                balance -= (balance * risk_per_trade)
                history.append({'time': df.index[i], 'res': 'SL', 'bal': balance, 'type': pos})
                pos = None; tapped = False; choch_found = False
            elif (pos == 'short' and row['low'] <= tp) or (pos == 'long' and row['high'] >= tp):
                balance += (balance * risk_per_trade) * 2
                history.append({'time': df.index[i], 'res': 'TP', 'bal': balance, 'type': pos})
                pos = None; tapped = False; choch_found = False
            continue

        # --- B. ENTRY LOGIC (PHASED) ---
        # 1. Tap: Price must touch or be inside the zone
        zone_s = row['priority_supply']
        zone_d = row['priority_demand']
        
        if not tapped:
            if (pd.notna(zone_s) and row['high'] >= zone_s): tapped = 'short'
            elif (pd.notna(zone_d) and row['low'] <= zone_d): tapped = 'long'
        
        # 2. Structural Confirmation (Swing break)
        swing_h = df['high'].iloc[max(0, i-10):i].max()
        swing_l = df['low'].iloc[max(0, i-10):i].min()
        
        if tapped == 'short' and not choch_found:
            if row['close'] < swing_l: choch_found = True
        elif tapped == 'long' and not choch_found:
            if row['close'] > swing_h: choch_found = True
            
        # 3. Final Entry: BOS (Lower structural break) + MACD
        macd_ok = (row['macd'] < row['macd_sig']) if tapped == 'short' else (row['macd'] > row['macd_sig'])
        
        if tapped == 'short' and choch_found and macd_ok:
            if row['close'] < prev['low']: # BOS confirmation
                pos = 'short'
                sl = zone_s + 0.5
                tp = row['close'] - (abs(sl - row['close']) * 2)
                
        elif tapped == 'long' and choch_found and macd_ok:
            if row['close'] > prev['high']: # BOS confirmation
                pos = 'long'
                sl = zone_d - 0.5
                tp = row['close'] + (abs(sl - row['close']) * 2)

    return pd.DataFrame(history)

# --- 4. MAIN FLOW ---
data_path = r'C:\Users\PREDATOR\Documents\BackTest\data'
df_1h = load_xau_file(os.path.join(data_path, 'XAU_1h_data.csv'))
df_15m = load_xau_file(os.path.join(data_path, 'XAU_15m_data.csv'))

# Process and Synchronize
h1_proc = identify_poi(df_1h, 'h1')
m15_proc = identify_poi(df_15m, 'm15')

# Manual MACD (1H)
ema12 = df_1h['close'].ewm(span=12, adjust=False).mean()
ema26 = df_1h['close'].ewm(span=26, adjust=False).mean()
h1_proc['macd'] = ema12 - ema26
h1_proc['macd_sig'] = h1_proc['macd'].ewm(span=9, adjust=False).mean()

# Merge onto execution TF (15m)
final = m15_proc.join(h1_proc[['active_supply_h1', 'active_demand_h1', 'macd', 'macd_sig']].shift(1))
final['priority_supply'] = final['active_supply_h1'].fillna(final['active_supply_m15'])
final['priority_demand'] = final['active_demand_h1'].fillna(final['active_demand_m15'])

# BACKTEST
results = run_backtest_pro(final)
if not results.empty:
    print(results['res'].value_counts())
    print(f"Final Balance: ${results['bal'].iloc[-1]:.2f}")
else:
    print("No trades triggered. Check POI/ChoCh sensitivity.")
```

```python
import pandas as pd
import numpy as np
import os

# --- 1. DATA LOADER ---
def load_xau_file(file_path):
    if not os.path.exists(file_path):
        print(f"Error: File {file_path} not found.")
        return pd.DataFrame()
    df = pd.read_csv(file_path, sep=';')
    df.columns = [col.strip().lower() for col in df.columns]
    df['date'] = pd.to_datetime(df['date'], format='%Y.%m.%d %H:%M')
    df.set_index('date', inplace=True)
    df.sort_index(inplace=True)
    return df

# --- 2. SMC LOGIC WITH TIME-SPECIFIC BCC ---
def identify_poi(df, tf_name):
    df = df.copy()
    
    # Core Candle Data
    df['body_size'] = (df['close'] - df['open']).abs()
    df['is_bullish'] = (df['close'] > df['open']).astype(bool)
    df['avg_body'] = df['body_size'].rolling(window=5).mean().fillna(0)
    df['is_impulsive'] = (df['body_size'] > (df['avg_body'] * 1.5)).astype(bool) # Slightly more sensitive

    # New Zone Detection
    df['new_supply'] = np.where((df['is_bullish'].shift(1) == True) & (df['is_bullish'] == False) & (df['is_impulsive'] == True), 
                                df['high'].shift(1), np.nan)
    df['new_demand'] = np.where((df['is_bullish'].shift(1) == False) & (df['is_bullish'] == True) & (df['is_impulsive'] == True), 
                                df['low'].shift(1), np.nan)

    # BCC Invalidation Loop (Timeframe Specific)
    active_s = np.nan
    active_d = np.nan
    s_history, d_history = [], []

    for i in range(len(df)):
        row = df.iloc[i]
        if pd.notna(row['new_supply']): active_s = row['new_supply']
        if pd.notna(row['new_demand']): active_d = row['new_demand']
        
        # BCC Rule: Only invalidate if the candle BODY closes past the zone
        if pd.notna(active_s) and row['close'] > active_s: active_s = np.nan
        if pd.notna(active_d) and row['close'] < active_d: active_d = np.nan
        
        s_history.append(active_s)
        d_history.append(active_d)

    df[f'active_supply_{tf_name}'] = s_history
    df[f'active_demand_{tf_name}'] = d_history
    return df

# --- 3. THE "STATE MACHINE" BACKTESTER ---
def run_full_backtest(df, initial_balance=10000, risk=0.01, buffer=0.3):
    balance = initial_balance
    pos, history = None, []
    
    # State tracking
    tapped, choch_found = False, False
    
    # Diagnostics
    diag = {'taps': 0, 'chochs': 0}

    for i in range(1, len(df)):
        row = df.iloc[i]
        prev = df.iloc[i-1]
        
        # A. EXIT LOGIC
        if pos:
            if (pos == 'short' and row['high'] >= sl) or (pos == 'long' and row['low'] <= sl):
                balance -= (balance * risk); pos = None; tapped = choch_found = False
                history.append({'time': df.index[i], 'res': 'SL', 'bal': balance})
            elif (pos == 'short' and row['low'] <= tp) or (pos == 'long' and row['high'] >= tp):
                balance += (balance * risk) * 2; pos = None; tapped = choch_found = False
                history.append({'time': df.index[i], 'res': 'TP', 'bal': balance})
            continue

        # B. ENTRY LOGIC
        zone_s = row['priority_supply']
        zone_d = row['priority_demand']

        # 1. Tap Phase (Price touches zone)
        if not tapped:
            if pd.notna(zone_s) and row['high'] >= zone_s: tapped = 'short'; diag['taps'] += 1
            elif pd.notna(zone_d) and row['low'] <= zone_d: tapped = 'long'; diag['taps'] += 1
        
        # 2. ChoCh Phase (Break of 3-candle swing)
        # Using a small lookback for speed in Gold
        sw_h = df['high'].iloc[max(0, i-3):i].max()
        sw_l = df['low'].iloc[max(0, i-3):i].min()

        if tapped == 'short' and not choch_found:
            if row['close'] < sw_l: choch_found = True; diag['chochs'] += 1
        elif tapped == 'long' and not choch_found:
            if row['close'] > sw_h: choch_found = True; diag['chochs'] += 1

        # 3. Final Entry: BOS + MACD
        macd_ok = (row['macd'] < row['macd_sig']) if tapped == 'short' else (row['macd'] > row['macd_sig'])
        
        if tapped and choch_found and macd_ok:
            if tapped == 'short' and row['close'] < prev['low']: # BOS
                pos = 'short'; sl = zone_s + buffer
                tp = row['close'] - (abs(sl - row['close']) * 2)
            elif tapped == 'long' and row['close'] > prev['high']: # BOS
                pos = 'long'; sl = zone_d - buffer
                tp = row['close'] + (abs(sl - row['close']) * 2)

    return pd.DataFrame(history), diag

# --- 4. MAIN FLOW ---
data_path = r'C:\Users\PREDATOR\Documents\BackTest\data'
df_1h = load_xau_file(os.path.join(data_path, 'XAU_1h_data.csv'))
df_30m = load_xau_file(os.path.join(data_path, 'XAU_30m_data.csv'))
df_15m = load_xau_file(os.path.join(data_path, 'XAU_15m_data.csv'))

# Process Timeframes
h1_p = identify_poi(df_1h, 'h1')
m30_p = identify_poi(df_30m, 'm30')
m15_p = identify_poi(df_15m, 'm15')

# Manual MACD (1H)
ema12 = h1_p['close'].ewm(span=12, adjust=False).mean()
ema26 = h1_p['close'].ewm(span=26, adjust=False).mean()
h1_p['macd'] = ema12 - ema26
h1_p['macd_sig'] = h1_p['macd'].ewm(span=9, adjust=False).mean()

# Merge onto 15m
final = m15_p.join(h1_p[['active_supply_h1', 'active_demand_h1', 'macd', 'macd_sig']].shift(1))
final = final.join(m30_p[['active_supply_m30', 'active_demand_m30']].shift(1))

# Priority Cascade (1H > 30m > 15m)
final['priority_supply'] = final['active_supply_h1'].fillna(final['active_supply_m30']).fillna(final['active_supply_m15'])
# The last part of the chain needs the '_m15' suffix
final['priority_demand'] = final['active_demand_h1'].fillna(final['active_demand_m30']).fillna(final['active_demand_m15'])

# Execute
results, d_report = run_full_backtest(final)

# OUTPUT
print("\n--- Diagnostic Report ---")
print(f"Total Taps Detected: {d_report['taps']}")
print(f"Total ChoChs Found:  {d_report['chochs']}")
print(f"Final Trades:        {len(results)}")
if not results.empty:
    wr = (results['res'] == 'TP').mean() * 100
    print(f"Win Rate: {wr:.2f}% | Final Balance: ${results['bal'].iloc[-1]:.2f}")
```
